## DATA VISUALIZATION WITH MATPLOTLIB
Data visualization is the graphical representation of
information and data. By using visual elements like charts,
graphs, and maps, data visualization tools provide an
accessible way to see and understand trends, outliers, and
patterns in data.

Matplotlib is one of the most popular Python packages used
for data visualization. It is a cross-platform library for
making 2D plots from data in arrays.

**matplotlib.pyplot** is a collection of command style functions that make Matplotlib work like MATLAB. Each Pyplot function makes some change to a figure. For example, a function creates a figure, a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x= np.arange(0, np.pi*2, 0.5)   #2*pi = 6.28
x
```

```python
y = np.sin(x)
y
#x is an independent variable, whereas y is a dependent(on x) variable
```

```python
#Univariate Analysis: The key fact is that only one variable is involved.
#Bivariate Analysis: Data analysis of two variables(usually two types of 
#related data)
#Multivariate Analysis: The statistical study of data where multiple 
#measurements are made on each experimental unit and where the relationships
#among multivariate measurements and their structure are important.
```

```python
#since there are less number of points plotted, graph does not appear smooth
plt.plot(x,y)
```

```python
x= np.arange(0, np.pi*2, 0.1) 
y= np.sin(x)
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine of Angle')
plt.title('Sine Function')
plt.show()  #this is not required in jupyter notebook because we are
#directly working in the environment therefore you can interact with the 
#memoru but in other IDE's where you are not working like this, you have to
#use this function explicitly to show the graph on the screen.
#NOTE: as you can see, graph appears smoother now.
```

```python
fig = plt.figure()
#plt.figure?
ax = fig.add_axes([0,0,1,1])  
#fig.add_axes?
#[rect]: left, bottom, width, height --> parameters for the graph window
#0,0 means the left bottom that is the edge of the window on bottom left, 
#1,1 here will give us the size of the axes to be completely in ratio (1:1)
#with the graph window
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.8,1.2])  
ax.plot(x,y)
ax.set_xlabel('Angle')
ax.set_ylabel('Sine of Angle')
ax.set_title('Sine Funcion')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.8,1.2]) 
ax.plot(x,np.cos(x))
ax.set_xlabel('Angle')
ax.set_ylabel('Cosine of Angle')
ax.set_title('Cosine Funcion')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.8,1.2]) 
ax.plot(x,np.power(x,8))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.8,1.2]) 
ax.plot(x,y, '--', color = 'DarkGreen')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.8,1.2]) 
ax.plot(x,y, '--s', color = 'Maroon')  #s: square marker
```

```python
#more than one plots in a single graphic window
fig = plt.figure()
ax = fig.add_axes([0,0,0.8,1.2]) 
ax.plot(x, np.cos(x), '--s', color = 'DarkGreen', label = 'Cosine')
ax.plot(x, np.sin(x), '-.*', color = '#A267Df', label = 'Sine')
ax.legend()  #matplotlib places the legend by automatically searching for 
#space available on the graph. using 'loc' argument you can dictate it's
#place.
```

# EXPLORATORY DATA ANALYSIS (EDA)
Exploratory Data Analysis(EDA) in Python is the first step to data analysis. In statistics, EDA is an approach to analyzing datasets to summarize their main characteristics, often with visual methods. By name itself, we can get to know that it is a step in which we need to explore the dataset.
- Feature engineering is a part of EDA.

### NEED for EDA:
EDA is  a crucial step before you jump to Machine Learning or modeling of your data. By doing this you can get to know whether the selected features are good enough to model, are all the features required, are there any  correlations based on which we can either go back to the Data Pre-processing step or move to the modeling.

Once EDA is complete and insights are drawn, it's feature can be used for supervised or unsupervised machine learning modeling.

In every machine learning workflow, the last step is Reporting or Providing the insights to the stake holders and as a data scientist you can explain every bit of code but you need to keep in mind the audience. By completing the Exploratory Data Analysis, you will have many plots, heat maps, frequency distribution graphs, correlation matrix along with the hypothesis by which any individual can understand what your data is all about and what insights you got from exploring your dataset.

**When you're building a machine learning model, you need to be pretty sure whether your data is making sense or not. The main aim of EDA is to obtain confidence in your data to an extent where you're ready to engage a machine learning algorithm.**

EDA is a phenomenon under data analysis used for gaining a better understanding of data aspects like:
- main features of data
- variables and relationships that hold between them
- identifying which variables are important for our problem

We shall look at various exploratory data analysis methods like:

- Descriptive Statistics, which is a way of giving a brief overview of the dataset we are dealing with, including some measures and features of the sample

- Grouping data [Basic grouping with group by]

- ANOVA, Analysis Of Variance, which is a computational method to divide variations in an observations set into different components.

- Correlation and correlation methods
