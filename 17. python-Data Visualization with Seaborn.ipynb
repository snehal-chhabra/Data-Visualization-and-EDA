## WHAT IS CORRELATION ANALYSIS?
- Correlation analysis in research is a statistical method used to measure the strength of the linear relationship between two variables and compute their association. Simply put, it calculates the level of change in one variable due to the change in the other.

```python
tc= titanic.corr(numeric_only = True)  #correlation matrix
tc
#titanic.corr?
#Compute pairwise correlation of columns, excluding NA/null values.
```

WHAT DO WE INFER FROM THIS TABLE?
-------
1. negative values tell us about an inversely proportional relationship here between two variables and positive values tell us about a directly  proportional relatinship.
2. The diagonal of the dataframe/matrix is 1 for every value because variability of category A with A is 1.
3. The triangular matrix above the diagonal is a mirror image of the lower triangular matrix. You can consider whichever you want.
4. For values that are very small, such as **-0.077221 between 'Survived' and 'Pclass'** or **-0.018443 between 'Pclass' and 'Parch'**, you can ignore those here if they don't provide much information.
5. The correlation between 'Survived' and 'Pclass' is **-0.338481** which is a significant value and cannot be ignored. It tells us **that higher the 'Pclass' value, lower will be the chances of survival.**

[**NOTE: numerically speaking, 3>2>1 which is the consideration made by the correlation matrix. When we say higher the 'Pclass' value we mean in literal number. But factually what we know is that higher the 'Pclass' , ie higher the socio-economical status/income/services bought, higher is the survival chances. Simply put, the passengers in 1st class paid more and bought better facilities and hence will be given more priority for survival than passengers in 3rd class.**]

6. Other correlations sch as 'Age' and 'SibSp' which amounts to **-0.308247** tells us that higher the age, less will be the number of siblings/spouses. It basically gives us an idea that the younger passengers are more likely to travel with their siblings/spouses than the older ones.
7. Another **important thing to be noted** is that correlation can only be evaluated betweeen numerical data therefore columns 'Name', 'Sex', 'Ticket', 'Cabin' and 'Embarked' are not taken into consideration. **If we need the correlation for any of these columns such as 'Embarked', we can convert it's data values into numeric format**, eg: S-> -1, C-> 0, Q-> 1, then find the correlation.
8. Observe the correlation **between 'SibSp' and 'Fare'** and the correlation **between 'Parch' and 'Fare'**. We note that they are both positive, therefore indicating that **higher the number of Siblings/Spouses/Parents/Children, higher is the fare.** Also, we note that the correlation value **between 'Parch' and 'Fare' is higher than the value between 'SibSp' and 'Fare'**. There could be many reasons behind it but you can make your own assumptions or evaluate firther if you have more data to work with.

## Violinplot using seaborn:
- Syntax: sns.violinplot(x,y)
   
        --> x: x-axis columns
        --> y: y-axis columns
- This plot helps you to analyze two columns, categorical vs continuous data. We can also plot a single column.
- It draws a combination of boxplot and kde.
- It will provide you the insights or patterns between them.

Draw a combination of boxplot and kernel density estimate.

A violin plot plays a similar role as a box and whisker plot. It shows the
distribution of quantitative data across several levels of one (or more)
categorical variables such that those distributions can be compared. Unlike
a box plot, in which all of the plot components correspond to actual
datapoints, the violin plot features a kernel density estimation of the
underlying distribution.

This can be an effective and attractive way to show multiple distributions
of data at once, but keep in mind that the estimation procedure is
influenced by the sample size, and violins for relatively small samples
might look misleadingly smooth.

```python
sns.violinplot?
```

```python
sns.violinplot(data = titanic, y= 'Age', x= 'Survived', hue= 'Sex')
#default of inner argument is box
```

```python
sns.violinplot(data = titanic, y= 'Age', x= 'Survived', inner ='quart', 
               hue= 'Sex', width = 0.5)
```

A strip plot can be drawn on its own, but it is also a good complement

## Swarmplot using seaborn:
- Syntax: sns.swarmplot(x,y)
   
        --> x: x-axis columns
        --> y: y-axis columns
- This plot helps you to analyze two columns, categorical vs continuous data. We can also plot a single column.
- It draws a scatterplot.
- It will provide you the insights or patterns between them.

Draw a categorical scatterplot with points adjusted to be non-overlapping.

This function is similar to :func: stripplot, but the points are adjusted
(only along the categorical axis) so that they don't overlap. This gives a
better representation of the distribution of values, but it does not scale
well to large numbers of observations. This style of plot is sometimes
called a "beeswarm".

A swarm plot can be drawn on its own, but it is also a good complement
to a box or violin plot in cases where you want to show all observations
along with some representation of the underlying distribution.

```python
sns.swarmplot?
```

```python
sns.swarmplot(data = titanic, x= 'Survived', y= 'Age', hue = 'Sex')
```

```python
sns.swarmplot(data = titanic, x= 'Survived', y= 'Age', hue = 'Sex', size = 3)
```

## Pairplot using seaborn:
- Syntax: sns.pairplot(data)

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric
variable in ``data`` will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated
differently: a univariate distribution plot is drawn to show the marginal
distribution of the data in each column.

It is also possible to show a subset of variables or plot different
variables on the rows and columns.

This is a high-level interface for :class:`PairGrid` that is intended to
make it easy to draw a few common styles. You should use :class:`PairGrid`
directly if you need more flexibility.

```python
sns.pairplot?
```

```python
sns.pairplot(titanic)
#this is like the correlation matrix for two columns at a time
```

```python
sns.pairplot(data= titanic, hue= 'Sex')
```

```python
sns.pairplot(data= titanic, hue= 'Sex', kind = 'reg')
#kind is by default scatter
```
