## Data Cleaning

```python
auto_copy = automobile.copy()
```

```python
auto_copy.isnull().sum()
```

### i) for 'normalized-losses' : 20% of data is missing. It is pointless to try and fill this data, and removing it would result in loss of other important datavalues. Therefore, we will replace null values with 'missing' in this column.

```python
auto_copy.fillna({'normalized-losses':'missing'}, inplace = True)
```

### ii) for 'num-of-doors' : only 2 data values are null, we will add the mode value of this column to replace null values.

```python
auto_copy['num-of-doors'].fillna(auto_copy['num-of-doors'].mode()[0], inplace = True)
```

### iii) for 'bore' and 'stroke' : we see that null values are present together and approx. 2% of data is missing for these columns. To fill them, we wil analyse their behaviour by first, converting them from 'object' datatype to 'float' datatype and then seeing their plots and correlations.

```python
auto_copy.loc[auto_copy['bore'].isnull()]   
```

```python
autobs_copy = auto_copy.copy()
#dropping missing values
autobs_copy.dropna(subset = ['bore','horsepower','peak-rpm','price'], inplace = True)
autobs_copy.isnull().sum()
```

```python
autobs_copy.sample(5)
```

```python
#CONVERTED TO float DATATYPE
autobs_copy[['bore','stroke']] = autobs_copy[['bore','stroke']].astype(float)
autobs_copy[['horsepower','peak-rpm','price']] = autobs_copy[['horsepower','peak-rpm','price']].astype(int)
autobs_copy.dtypes
```

```python
autobs_copy.corr(numeric_only = True)
```

```python
sns.displot(x= autobs_copy['bore'])
```

```python
sns.boxplot(x= autobs_copy['bore'])
```

```python
sns.jointplot(x= autobs_copy['bore'], y= autobs_copy['curb-weight'])
```

```python
sns.displot(x= autobs_copy['stroke'])
```

```python
sns.boxplot(x= autobs_copy['stroke'])
```

## OBSERVATION--> 
- **'bore' has intermediate correlation with 'curb-weight'. On analyzing their plots together using a jointplot, and on analyzing boxplot for 'bore', we observe that for missing data, we can replace them with the median value of bore data.**
- **'stroke' does not have significant correlation with any other column. But it's boxplot shows an almost perfect distribution which allows us to replace the missing data with it's median value.**

```python
auto_copy['bore'].fillna(autobs_copy['bore'].median(), inplace = True)
auto_copy['stroke'].fillna(autobs_copy['stroke'].median(), inplace = True)
```

```python
auto_copy.loc[55:58]
```
